# Japanese Survey Analysis

Data Source: [Survery Research Center](https://www.surece.co.jp/), [JAG Japan](https://gis.jag-japan.com/covid19jp/)


```python
import pandas as pd
import matplotlib.pyplot as plt

def read_survey_results():
    df = pd.read_csv('data/SurveyResearchCenter/data.csv', na_values='@')
    # print(df['Q5-1'])
    df = df.groupby("Q1").mean()
    df = df.drop(columns=["ID", "Q2.1", "Q2.2", "Q2.3", "Q2.4", "Q2.5", "Q2.6", "Q2.7", "Q2.8", "Q3", "NQ3", "F1", "F2", "NF2"])
    pref_name = pd.read_csv('data/SurveyResearchCenter/pref_code.csv', header=None, index_col=0)
    df = df.assign(SAMPLENUMBER=pref_name)
    df = df.set_index("SAMPLENUMBER")
    return df

def select_query(q):
    c = df.columns
    queries = c[c == q]
    if len(queries) == 0:
        queries = c[c.str.startswith(q)]
    return queries

# print(df['Q5-1'])

def doit():
    for q_no in range(3, 37):
        q = 'Q' + str(q_no)
        df[select_query(q)].boxplot()
        plt.show()

# doit()

survey_results = read_survey_results()
survey_results
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# fn = 'https://dl.dropboxusercontent.com/s/6mztoeb6xf78g5w/COVID-19.csv'
fn = 'data/JAG_Japan/COVID-19.csv'

def confirmed_by_date(date):
    # fn = 'https://dl.dropboxusercontent.com/s/6mztoeb6xf78g5w/COVID-19.csv'
    fn = 'data/JAG_Japan/COVID-19.csv'
    d = pd.read_csv(fn,usecols=["確定日", "受診都道府県"],parse_dates=["確定日"])
    d = d[d["確定日"] < date].groupby("受診都道府県").count().rename(columns={"確定日": date})
    d = d.drop(["不明", "中部国際空港", "成田空港", "羽田空港", "関西国際空港"], errors='ignore')
    return d

# pd.to_datetime("2020/3/6").date()
before_date = "2020/3/6"
after_date = "2020/3/27"
df = survey_results
df = df.join(confirmed_by_date(before_date))
df = df.join(confirmed_by_date(after_date))
# confirmed_by_date("2020/3/6")
df["Ratio"] = df[after_date] / df[before_date]
corr = df.fillna(value=0).corr().drop(labels=[before_date, after_date, "Ratio"])

def show_highcorr(df, key, fn):
    df = df.sort_values(key, ascending=False)
    print("Most positive five on {}".format(key))
    display(df.iloc[:5,:][key])
    df.iloc[:5,:][key].to_csv("SRC_{}.csv".format(fn), mode="w")
    print("Most negative five on {}".format(fn))
    display(df.iloc[-5:,:][key])
    df.iloc[-5:,:][key].to_csv("SRC_{}.csv".format(fn), mode="a")
    return

show_highcorr(corr, before_date, "before")
show_highcorr(corr, after_date, "after")
show_highcorr(corr, "Ratio", "Ratio")
corr
```

```python
# 確定症例報告数を都道府県マップ上に可視化
import japanmap
%config InlineBackend.figure_formats = {'png', 'retina'}
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
# df = pd.read_excel('data/istats/02-01.xls', usecols=[2, 3], skiprows=3, skipfooter=1,
#                   names='都道府県 人口'.split()).set_index('都道府県')
# df = df.drop("全国")


def pref_map(df, title="", max=300):
    cmap = plt.get_cmap('Reds')
    # norm = plt.Normalize(vmin=df.min(), vmax=df.max())
    norm = plt.Normalize(vmin=df.min(), vmax=max)
    fcol = lambda x: '#' + bytes(cmap(norm(x), bytes=True)[:3]).hex()
    plt.title(title)
    plt.colorbar(plt.cm.ScalarMappable(norm, cmap))
    plt.imshow(japanmap.picture(df.apply(fcol)))
    plt.show()
    return

df = confirmed_by_date(before_date)
pref_map(df[before_date], title="Confirmed cases on {}".format(before_date))

df = confirmed_by_date(after_date)
pref_map(df[after_date], title="Confirmed cases on {}".format(after_date))

df = confirmed_by_date(after_date).join(confirmed_by_date(before_date))
df['Ratio'] = df[after_date] / df[before_date]
pref_map(df['Ratio'].fillna(value=0), title="Growth rate of confirmed cases", max=10)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%load_ext autotime

year = "2020"

fn = "data/DoCoMo/pattern07_mesh_pop_20190101-20190114/ordermade_sokuho_20190101_pattern07_mesh_pop.csv"
fn = "data/DoCoMo/" + year + "_all.csv"
d = pd.read_csv(fn,usecols=[0,1,2,6],names=["Date", "Time", "Mesh", "Population"],
                # dtype={"Mesh":str},
                parse_dates=[[0,1]],header=None)
d.set_index("Mesh", inplace=True)
# print(d)
print("n_grids=", len(d.groupby("Mesh").count().index))
# d.groupby("Mesh").count()
# pd.DataFrame(d["Population"].values.reshape(6269, 24), index=d["Mesh"].values)
start_time, end_time = d["Date_Time"].min(), d["Date_Time"].max()
print("time interval=", start_time, end_time)
# r = d.loc[374112352]
# d[d["Mesh"]==374112352]
mesh_list = d.index.unique()
print("mesh_list=", mesh_list)
# a = d.loc[mesh_list[0], ["Date_Time","Population"]]
# time_series = pd.DataFrame(a,index = mesh_list)
# print("a=", a)
# print("time_series=", time_series)
# print("===", type(d.loc[mesh_list[2]]))

#print("===", type(d.loc[mesh_list[2]]))
#print(d.loc[mesh_list[2]].reset_index().drop(columns=["Mesh"]))
#print("===", type(d.loc[mesh_list[8]]))
#print(pd.DataFrame(d.loc[mesh_list[8],:]))
#print(pd.DataFrame(d.loc[mesh_list[8],:]).T)
# print(d.loc[mesh_list[8],:].T)
#print("===")
time_series = d.loc[mesh_list[0]] \
        .reset_index() \
        .set_index("Date_Time") \
        .drop(columns=["Mesh"]) \
        .rename(columns={"Population":mesh_list[0]})
# print(time_series)
for i in range(1, len(mesh_list)):
    if i % 100 == 0:
        print("{}: Processing grid {} ...".format(pd.Timestamp.now(), i))
    mesh = mesh_list[i]
    # print(mesh)
    a = d.loc[mesh]
    # When a single row matches, d.loc[mesh] returns a Series instead of DataFrame
    # print(mesh, len (a))
    if type(a) == pd.core.frame.DataFrame:
        a = a.reset_index()
        a = a.drop(columns=["Mesh"])
    else:
        a = pd.DataFrame(a).T
    time_series = time_series.merge(a, on="Date_Time", how='outer') \
        .rename(columns={"Population":mesh})

# r.reset_index().drop(columns=["Mesh"]).set_index("Date_Time").plot()
time_series = time_series.set_index("Date_Time").fillna(0)
time_series.to_csv("Data/DoCoMo/" + year + "_out.csv")
time_series
```
